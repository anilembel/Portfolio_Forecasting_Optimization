## Scraping Data

### Adj Close

```python

# I want to get data on the football clubs Borussia Dortmund, Real Madrid, Galatasaray and Juventus
symbols = ['BVB.DE', 'AJAX.AS', 'OLG.PA', 'JUVE.MI']

#  I used yahoo finance to get the data
start_date = '2018-01-01'
end_date = datetime.today().strftime('%Y-%m-%d')

data1 = yf.download(symbols[0], start=start_date, end=end_date)['Adj Close']
data2 = yf.download(symbols[1], start=start_date, end=end_date)['Adj Close']
data3 = yf.download(symbols[2], start=start_date, end=end_date)['Adj Close']
data4 = yf.download(symbols[3], start=start_date, end=end_date)['Adj Close']

data_Adj = pd.concat([data1, data2, data3, data4], axis=1)
```

```python
data_Adj.info()
```

```python
data_Adj.dropna()
```

```python
data_Adj.to_csv('C:\\Users\\pc\\Desktop\\Studies\\BECODE\\Projects\\Portfolio_Forecasting_Optimization\\datas\\data_Adj.csv')
```

```python
column_names = ['Adj Close'] * len(symbols)

new_column_names = [f"{symbol}_{col}" for symbol, col in zip(symbols, column_names)]

data_Adj.columns = new_column_names
```

```python
# GÃ¼nlÃ¼k getirileri hesaplayÄ±nÄ±z
daily_returns = data_Adj.pct_change()

# Log getirileri hesaplayÄ±nÄ±z
log_returns = np.log(data_Adj / data_Adj.shift(1))

# GÃ¼nlÃ¼k getiri ortalamanÄ±zÄ± hesaplayÄ±nÄ±z
mean_returns = daily_returns.mean()

# GÃ¼nlÃ¼k getiri kovaryans matrisini hesaplayÄ±nÄ±z
cov_matrix = daily_returns.cov()

# GÃ¼nlÃ¼k getirilerin volatilitesini hesaplayÄ±nÄ±z
volatilities = daily_returns.std()

# Sharpe oranÄ±nÄ± hesaplayÄ±nÄ±z
risk_free_rate = 0.02
sharpe_ratios = (mean_returns - risk_free_rate) / volatilities

# SonuÃ§larÄ± yazdÄ±rÄ±nÄ±z
print("GÃ¼nlÃ¼k getiriler:")
print(daily_returns.tail())

```
